```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
from pandas.plotting import scatter_matrix
```

```python
dataEURUSD = pd.read_csv('EURUSD.csv')
pd.concat([dataEURUSD.head(), dataEURUSD.tail()])
```

```python
dataUSDJPY = pd.read_csv('USDJPY.csv')
pd.concat([dataUSDJPY.head(), dataUSDJPY.tail()])
```

```python
dataGBPUSD = pd.read_csv('GBPUSD.csv')
pd.concat([dataGBPUSD.head(), dataGBPUSD.tail()])
```

```python
dataUSDCHF = pd.read_csv('USDCHF.csv')
pd.concat([dataUSDCHF.head(), dataUSDCHF.tail()])
```

```python
dataEURUSD = pd.read_csv('EURUSD.csv')
dataEURUSD = dataEURUSD.dropna()
dataUSDJPY = pd.read_csv('USDJPY.csv')
dataUSDJPY = dataUSDJPY.dropna()
dataGBPUSD = pd.read_csv('GBPUSD.csv')
dataGBPUSD = dataGBPUSD.dropna()
dataUSDCHF = pd.read_csv('USDCHF.csv')
dataUSDCHF = dataUSDCHF.dropna()
```

```python
close_px_EURUSD = dataEURUSD['Close']
mavg = close_px_EURUSD.rolling(window=100).mean()

fig = plt.gcf()
fig.set_size_inches(15,10)
close_px_EURUSD.plot(label='EUR/USD')
mavg.plot(label='100 MA')
plt.ylabel('PRICE',fontsize=15)
plt.xticks(range(0,dataEURUSD.shape[0],500),dataEURUSD['Date'].loc[::500],rotation=45)
plt.xlabel('DATE',fontsize=15)
plt.title('USD/JPY Multi-Sequence Prediction',fontsize=18)
plt.legend()
```

```python
close_px_USDJPY = dataUSDJPY['Close']
mavg = close_px_USDJPY.rolling(window=100).mean()

fig = plt.gcf()
fig.set_size_inches(15,10)
close_px_USDJPY.plot(label='USD/JPY')
mavg.plot(label='100 MA')
plt.ylabel('PRICE',fontsize=15)
plt.xticks(range(0,dataUSDJPY.shape[0],500),dataUSDJPY['Date'].loc[::500],rotation=45)
plt.xlabel('DATE',fontsize=15)
plt.legend()
```

```python
close_px_GBPUSD = dataGBPUSD['Close']
mavg = close_px_GBPUSD.rolling(window=100).mean()

fig = plt.gcf()
fig.set_size_inches(15,10)
close_px_GBPUSD.plot(label='GBP/USD')
mavg.plot(label='100 MA')
plt.ylabel('PRICE',fontsize=15)
plt.xticks(range(0,dataGBPUSD.shape[0],500),dataGBPUSD['Date'].loc[::500],rotation=45)
plt.xlabel('DATE',fontsize=15)
plt.legend()
```

```python
close_px_USDCHF = dataUSDCHF['Close']
mavg = close_px_USDCHF.rolling(window=100).mean()

fig = plt.gcf()
fig.set_size_inches(15,10)
close_px_USDCHF.plot(label='USD/CHF')
mavg.plot(label='100 MA')
plt.ylabel('PRICE',fontsize=15)
plt.xticks(range(0,dataUSDCHF.shape[0],500),dataUSDCHF['Date'].loc[::500],rotation=45)
plt.xlabel('DATE',fontsize=15)
plt.legend()
```

```python
rets = close_px_EURUSD / close_px_EURUSD.shift(1) - 1
fig = plt.gcf()
fig.set_size_inches(15,10)
rets.plot(label='return')
plt.ylabel('RETURN RATE',fontsize=15)
plt.xticks(range(0,dataEURUSD.shape[0],500),dataEURUSD['Date'].loc[::500],rotation=45)
plt.xlabel('DATE',fontsize=15)
```

```python
rets = close_px_USDJPY / close_px_USDJPY.shift(1) - 1
fig = plt.gcf()
fig.set_size_inches(15,10)
rets.plot(label='return')
plt.ylabel('RETURN RATE',fontsize=15)
plt.xticks(range(0,dataUSDJPY.shape[0],500),dataUSDJPY['Date'].loc[::500],rotation=45)
plt.xlabel('DATE',fontsize=15)
```

```python
rets = close_px_GBPUSD / close_px_GBPUSD.shift(1) - 1
fig = plt.gcf()
fig.set_size_inches(15,10)
rets.plot(label='return')
plt.ylabel('RETURN RATE',fontsize=15)
plt.xticks(range(0,dataGBPUSD.shape[0],500),dataGBPUSD['Date'].loc[::500],rotation=45)
plt.xlabel('DATE',fontsize=15)
```

```python
rets = close_px_USDCHF / close_px_USDCHF.shift(1) - 1
fig = plt.gcf()
fig.set_size_inches(15,10)
rets.plot(label='return')
plt.ylabel('RETURN RATE',fontsize=15)
plt.xticks(range(0,dataUSDCHF.shape[0],500),dataUSDCHF['Date'].loc[::500],rotation=45)
plt.xlabel('DATE',fontsize=15)
```

```python
dataComp0= pd.merge(dataEURUSD,dataUSDJPY, left_on=['Date'], right_on=['Date'])
dataComp1 = pd.merge(dataComp0, dataGBPUSD,left_on=['Date'], right_on=['Date'])
dataComp = pd.merge(dataComp1, dataUSDCHF,left_on=['Date'], right_on=['Date'])
dataComp =  dataComp.drop('Date', 1)
dataComp.columns = ['EURUSD', 'USDJPY', 'GBPUSD', 'USDCHF']
dataComp.head()
```

```python
retscomp = dataComp.pct_change()
corr = retscomp.corr()
corr
```

```python
fig = plt.gcf()
fig.set_size_inches(15,10)
plt.scatter(retscomp.EURUSD, retscomp.USDCHF)
plt.xlabel('Returns EURUSD')
plt.ylabel('Returns USDCHF')
```

```python
fig = plt.gcf()
fig.set_size_inches(15,10)
plt.scatter(retscomp.EURUSD, retscomp.GBPUSD)
plt.xlabel('Returns EURUSD')
plt.ylabel('Returns GBPUSD')
```

```python
pd.plotting.scatter_matrix(retscomp, diagonal='kde', figsize=(15, 15))
```

```python
fig = plt.gcf()
fig.set_size_inches(15,10)
plt.imshow(corr, cmap='hot', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns)
plt.show
```
